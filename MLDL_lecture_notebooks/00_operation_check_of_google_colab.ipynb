## What's Google Colaboratory?

Google Colaboratory (Google Colab.) is the cloud-based Jupyter Notebook environment provided by Google Inc.
Google Colab. is developed for the purpose of education and research of machine learning or image recognition.
We can use this service for free.

## Google Colab. (Jupyter Notebook) Cell

In Google Colab., we arrange programs by writing programs or any descriptions in each *cell*.
The cell is categorized by the following two types:
1. Markdown cell and
2. Code cell.


#### 1. Markdown Cell
Markdown Cell write any descriptions and equations by Markdown syntax that is used in Github README, etc.

#### 2. Code Cell
Code cell write Python program.
As the ordinary Python program, we can write any programs.


## How to execute programs?

By clicking execution button placed at the left of code cell, we can run each code cell one by one.
In the another way, we can run by pressing `Shift + Enter` while the code cell is selected.
